## Estimating the Incumbency Effect

Everyone talks about anti-incumbency being dead, but no one brings the evidence. We're looking out to identify if there's any basis for this axiom. We consider all State Assembly Elections for current states since 1962. Our data is from the Trivedi Centre for Political Data. We try estimating if incumbency is an important predictor for victory or not. 

```python
#Importing all packages relevant
import pandas as pd
from plotnine import *
import numpy as np
import glob as glob
import os
# importing statsmodels (for linear regression in python)
import statsmodels.formula.api as smf
pd.set_option('display.max_columns', None)
```

```python
# path = r'C:\\Users\\sriha\\Desktop\\Columbia\\spring2022\\datacomputationinnovation\\incumbency-effect\\data' # use your path
all_files = glob.glob(("*.csv"))

# df = pd.concat((pd.read_csv(f) for f in all_files))
```

```python
#We run through all elections for each state, for each election we look at whether incumbency explains the results, and attempt a regression. 
data = []
for file in all_files:
        df = pd.read_csv(file)
        list_of_all_years=df.Year.unique()
        list_of_assembly_elections=[]
        for year in list_of_all_years:
            tempdf=df[df['Year']==year]
            if tempdf.Year.count() > 50:
                list_of_assembly_elections.append(year)
                try: 
                    elections = tempdf
                    elections['Incumbency_Value'] = np.where(elections['Incumbent'] == 'False', 0, elections['Incumbent'])
                    elections['Winner'] = np.where(elections['Position'] == 1, 1,0)
                    model = smf.logit("Winner ~ No_Terms + Recontest + N_Cand + Incumbent + ENOP", data=elections) #Our regression
                    results = model.fit()
                    x = results.params
                    y = results.pvalues
                    coefs = {"coef_" + str(key): val for key, val in x.items()}
                    pvalues =  {"p_" + str(key): val for key, val in y.items()}
                    finaldict ={**coefs, **pvalues}
                    finaldict['year'] =year
                    finaldict['state'] = tempdf.State_Name.value_counts().index[0]
                    data.append(finaldict)
                except:
                    pass
                
dataframe = pd.DataFrame(data)
```

```python
#We exclude all results where incumbency was not a signifiant factor in the results ie
newdf=dataframe[dataframe['p_Incumbent[T.True]'] < 0.05]
```

```python
#We make a datafarme with the coefficient of Incumbency and other variables
newdf.head()
```

```python
#We display the coefficient of incumbency across states
display(
    ggplot(newdf,                   # data
        aes('year', 'coef_Incumbent[T.True]'))          # aesthetics
        + geom_line()               # geometry
        + facet_wrap('state')
        + theme(figure_size=(25, 16 ))
        + labs(
            title = "States",
            y = "coef_fit",
            x = "year"
        )
)   
```

## Estimating what proportion of winners are incumbents across time

```python
#We run the code for all files and calculate how many winners in every election across state and time are incumbents or not
all_the_data=[]
for file in all_files:
        df = pd.read_csv(file)
        list_of_all_years=df.Year.unique()
        for year in list_of_all_years:
            each_row={}
            tempdf=df[df['Year']==year]
            each_row['year'] = year
            each_row['state'] = tempdf['State_Name'].unique()[0]
            each_row['no_constituencies'] = len(tempdf['Constituency_Name'].unique())
            each_row['no_candidates'] = len(tempdf['pid'].unique())
            incumbentdf=tempdf[tempdf['Incumbent']==True]
            each_row['no_incumbents_participating'] = len(incumbentdf)
            each_row['no_incumbents_winning'] = len(incumbentdf[incumbentdf['Position']==1])
            all_the_data.append(each_row)
```

```python
#Making it a dataframe
incumbentdf= pd.DataFrame(all_the_data)
```

```python
#Calculating the percentage of winners who are incumbents
incumbentdf['percentage_winner'] = round(incumbentdf['no_incumbents_winning']*100/incumbentdf['no_constituencies'],1)
```
